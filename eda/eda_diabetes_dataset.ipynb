## Feature Engineering

```python
X = pd.DataFrame()

X['age'] = data.age
X['is_male'] = data.gender.apply(lambda x: 1 if x == 'Male' else 0)

for attr in data.columns.drop(['age', 'gender', 'class']):
    X[attr] = data[attr].apply(lambda x: 1 if x == 'Yes' else 0)
```

```python
AGE_RANGES = {
    56: '56_xx',
    46: '46_55',
    36: '36_45',
    26: '26_35',
    16: '16_25'
}

def categorize_age(age):
    for upper_bound, label in AGE_RANGES.items():
        if age >= upper_bound:
            return label

age_one_hot = pd.get_dummies(data.age.apply(categorize_age), drop_first=True)
X = pd.merge(X, age_one_hot.add_prefix('age_'), how='left', left_index=True, right_index=True)
```

```python
print(X.shape)
X.head()
```

```python
# Assign target variable
y = data['class'].apply(lambda x: 1 if x == 'Positive' else 0).rename('is_diabetic')
print(y.shape)
print(y.value_counts())
```

### Save transformed data to CSVs

```python
X.to_csv('../matrices/diabetes/diabetes_X.csv')
y.to_csv('../matrices/diabetes/diabetes_y.csv')
```
